```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

kashti = sns.load_dataset("titanic")
kashti.head()
sns.boxplot(x= "survived", y="age", data=kashti)
plt.show()
```

```python
sns.boxplot(x= "survived", y="age",showmeans=True,
            meanprops= {"marker": "+",
                       "markersize": "12",
                       "markeredgecolor": "red"},
            data=kashti)
#show labels
plt.xlabel("how many survived", size= 14),
plt.ylabel("age in years", size= 30),
plt.title("da da da doba man darya main", size= 15, weight="bold")
plt.show()
```
